```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=open('isoentropico.dat','r').readlines()
data=list(np.transpose([np.asarray(x.split(),np.float64) for x in data]))
```

```python
fig=plt.figure(figsize=(6,6),dpi=300)
ax=fig.add_subplot(111)
ax.plot(data[0],data[1],label=r'$p/p_0$')
ax.plot(data[0],data[2],label=r'$\rho/\rho_0$')
ax.plot(data[0],data[3],label=r'$T/T_0$')
ax.plot(data[0],data[4],label=r'$A/A^*$')
plt.legend()
```
