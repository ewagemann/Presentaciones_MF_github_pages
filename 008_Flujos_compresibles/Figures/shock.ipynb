```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(8,6),dpi=300)
ax=fig.add_subplot(111)
Ma=np.linspace(1,5,1000)
k=1.4


rho=np.divide(np.multiply((k+1),np.power(Ma,2)),np.multiply(k-1,np.power(Ma,2))+2)

T=np.multiply(2+np.multiply(k-1,np.power(Ma,2)),np.divide(np.multiply(2*k,np.power(Ma,2))-(k-1),\
                                                          np.multiply((k+1)**2,np.power(Ma,2))))

Ma2=np.divide(np.multiply(k-1,np.power(Ma,2))+2,np.multiply(2*k,np.power(Ma,2))-(k-1))
Ma2=np.sqrt(Ma2)

p02p01=np.power(np.divide(np.multiply(k+1,np.power(Ma,2)),2+np.multiply(k-1,np.power(Ma,2))),(k/(k-1)))
p02p01=np.multiply(p02p01,np.power(np.divide(k+1,np.multiply(2*k,np.power(Ma,2))-(k-1)),1/(k-1)))

T0=1+0.2*np.power(Ma,2)
p2p1=np.multiply(1/(k+1),np.multiply(2*k,np.power(Ma,2))-(k-1))

A2A1=np.divide(2+np.multiply(k-1,np.power(Ma,2)),2+np.multiply(k-1,np.power(Ma2,2)))
A2A1=np.power(A2A1,0.5*(k+1)/(k-1))
A2A1=np.multiply(A2A1,np.divide(Ma2,Ma))

ax.plot(Ma,Ma2,label=r'$Ma_2$')
ax.plot(Ma,T,label=r'$T_2/T_1$')
ax.plot(Ma,rho,label=r'${\rho_2}/{\rho_1}=V_1/V_2$')
ax.plot(Ma,p2p1,label=r'$p_2/p_1$')
ax.plot(Ma,p02p01,label=r'$p_{0_2}/p_{0_1}=\rho_{0_2}/\rho_{0_1}$')
ax.plot(Ma,A2A1,label=r'$A_2^*/A_1^*$')

ax.set_xticks(np.arange(1,5.5,0.5))
ax.set_yticks(np.arange(0,6.5,0.5))

ax.set_xlim(1,5)
#plt.yscale('log')
ax.set_ylim(0,6)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.grid('True')
#ax.legend(,loc='lower right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=15)
ax.set_xlabel('Ma$_1$',fontsize=20)
plt.tight_layout()
plt.savefig('shock.eps')
plt.savefig('shock.png')
```
